```python
# score data
def create_score_df():
    
    # Join Assessment and StudentAssessment tables
    joined=pd.merge(stud_asses_df,assessment_df_fff,on='id_assessment',how='left')
    
    # Calculate weighted scores for all assessments of all students
    joined['score*weight']=(joined['score']*joined['weight'])

    # Sum up score*weights and divide by total weights
    marks = joined.groupby(['id_student','module_presentation'],as_index=False)['score*weight','weight'].sum()
    
    # vēl var izveidot weighted score dalot scor*weight/weight, bet tas īsti nav būtiski
    # jo tad studentam, kurš labi izpildījis vienu assignment gala rezultātā sanāks final score tikai tā assignment rezultāts
    # kas nav loģiski, jo ir jāņem vērā, ka kopējais visu assignment weight ir 100
    
    # 100 ir visu assignmentu weight sum
    marks['w_final_score'] = marks['score*weight']/100

    marks = marks.round(1)
    return marks
```

```python
# no https://www.kaggle.com/code/zincbottom/oulad-random-forest Prepping the Data for Analysis

# function for merging 2 pandas dataframes

def inner_merge(left_df,right_df,right_cols,on_cols):
    right_df = right_df[right_cols]
    left_df = left_df.merge(right_df, on = on_cols)
    return left_df.drop_duplicates()
```

```python
# click data

def create_click_df():

    clicks = stud_vle_df_fff

    clicks = inner_merge(clicks,
                         vle_df_fff,
                         ['id_site','module_presentation','activity_type'],
                         ['id_site','module_presentation'],
                        )

    clicks = clicks.drop('id_site', axis = 1)
    #print(clicks)

    def clicks_xx(clicks):
        temp = clicks
        temp = temp.drop('date', axis = 1)
        temp = temp.groupby(['module_presentation','id_student','activity_type']).sum()
        temp = temp.reset_index() # this fills out the missing columns for merging later 
        return temp
    
    #print(clicks_xx(clicks))

    click_data = pd.pivot_table(data = clicks_xx(clicks), 
                                index = ['module_presentation','id_student'],
                                columns = 'activity_type', 
                                values = ['sum_click'],
                                fill_value = 0,
                               ).reset_index()

    # get rid of multi index
    click_data = pd.concat([click_data['module_presentation'],
                            click_data['id_student'],
                            click_data['sum_click']], axis=1)
    return click_data
```

```python
# merge score_df and click_df
def create_final_df(score_df, click_df):

    final_df = inner_merge(click_df, # merge with StudentInfo
                           stud_info_df_fff,
                           stud_info_df_fff.columns,
                           ['module_presentation', 'id_student'],
                          )

    final_df = final_df.replace('Distinction','Pass') # merge Pass and Distinction

    final_df = inner_merge(final_df, 
                           stud_reg_df_fff,
                           ['module_presentation', 'id_student', 'date_unregistration'],
                           ['module_presentation', 'id_student'],
                          )

    final_df = final_df[(final_df['final_result'] != 'Withdrawn')] # remove people who withdrew

    final_df = final_df.reset_index()

    final_df = final_df.drop(['date_unregistration','index'],axis = 1)

    final_df = inner_merge(final_df,
                           score_df,
                           score_df.columns,
                           ['module_presentation', 'id_student'])

    return final_df
```

```python
score_df = create_score_df()
click_df = create_click_df()
final_df = create_final_df(score_df, click_df)
final_df
```
