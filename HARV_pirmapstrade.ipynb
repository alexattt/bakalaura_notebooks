```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
harv_df = pd.read_csv("HXPC13_DI_v3_11-13-2019.csv")
print(harv_df.shape)
harv_df.head(10)
```

```python
harv_df.describe()
```

```python
# nav būtiska, visiem studentiem vērtība ir NaN
harv_df = harv_df.drop(columns=['roles'])
harv_df.head()
```

```python
print(harv_df.course_id.describe())
print(harv_df.course_id.unique())
print(harv_df.course_id.value_counts())
```

```python
print(harv_df.LoE_DI.describe())
print(harv_df.LoE_DI.unique())
```

```python
# formatting start_time_DI and last_event_DI columns to a standard date form
harv_df[["start_time_DI", "last_event_DI"]] = harv_df[["start_time_DI", "last_event_DI"]].apply(pd.to_datetime)
harv_df.head()
```

```python
# getting Nan count in each column
harv_df.isnull().sum(axis = 0)
```

```python
# nav nepieciešami datu objekti, kur gender vai YoB ir NaN
harv_df = harv_df.dropna(subset=['gender', 'YoB'])
print(harv_df.shape)
harv_df.head()
```

```python
# replacing all the NaN's in education column with Unknown
harv_df.LoE_DI = harv_df.LoE_DI.fillna('Unknown')
harv_df.head()
```

```python
# ja last_event_DI ir NaN, students pēc reģistrēšanās kursā, vairs nav to apmeklējis
harv_df = harv_df.dropna(subset=['last_event_DI'])
print(harv_df.shape)
harv_df.head()
```

```python
harv_df.nevents = harv_df.nevents.fillna(0)
harv_df.ndays_act = harv_df.ndays_act.fillna(0)
harv_df.nplay_video = harv_df.nplay_video.fillna(0)
harv_df.nchapters = harv_df.nchapters.fillna(0)
harv_df.nforum_posts = harv_df.nforum_posts.fillna(0)
harv_df.grade = harv_df.grade.fillna(0)
```

```python
# lietotājs ir certified un passing course, ja grade šajā priekšmetā ir >=0.5
harv_df_ph = harv_df.loc[harv_df['course_id'] == 'HarvardX/PH278x/2013_Spring']
print(harv_df_ph.shape)
harv_df_ph.head()
```

```python
harv_df_ph = harv_df_ph.drop(columns=['registered', 'viewed', 'explored', 'final_cc_cname_DI', 'start_time_DI', 'last_event_DI', 'incomplete_flag'])
print(harv_df_ph.shape)
harv_df_ph.head()
```

```python
harv_df_ph = harv_df_ph.drop(columns=['course_id', 'userid_DI'])
print(harv_df_ph.shape)
harv_df_ph.head()
```

```python
harv_df_ph['LoE_DI'] = harv_df_ph['LoE_DI'].str.replace("'",'')
print(harv_df_ph.shape)
harv_df_ph.head()
```

```python
harv_ph_cert= harv_df_ph.loc[harv_df_ph['certified'] == 1]
print(harv_ph_cert.shape)
harv_ph_cert.head()
```

```python
# creating new modification of harv_df_ph, where I am getting rid of grade and certified columns
# and creating final_mark instead, with two categories - Fail and Pass

harv_ph = harv_df_ph
harv_ph.head()
```

```python
def map_cert(certified):
  if certified == 1:
    return "Pass"
  else:
    return "Fail"

harv_ph["final_mark"] = harv_ph["certified"].apply(lambda cert: map_cert(cert))
harv_ph.head()
```

```python
harv_ph.final_mark.value_counts()
```

```python
harv_ph = harv_ph.drop(columns=['certified', 'grade'])
harv_ph.head()
```

```python
# harv_ph.to_csv('HARV_PH278_preprocessed.csv', index=False)
```

```python
# tiek veidota kategorizēta datu kopa
harv_ph_catg = harv_ph

# vecumu aprēķina atņemot kursa norises gadu 2013 no YoB

harv_ph_catg.rename(columns={'YoB':'age'}, inplace=True)
harv_ph_catg['age'] = 2013 - harv_ph_catg['age']

harv_ph_catg.head()
```

#### turpmākajos soļos notiek age, nevents, nplay_video, nchapters, ndays_act kategorizēšana

```python
def map_age(age):
  if age >= 0 and age < 35 :
    return "0-35"
  elif age >= 35 and age < 55:
    return "35-55"
  else:
    return "55<="

harv_ph_catg["age"] = harv_ph_catg["age"].apply(lambda age: map_age(age))
harv_ph_catg.head()
```

```python
def map_events(events):
  if events >= 0 and events < 100:
    return "0-100"
  elif events >= 100 and events < 1000:
    return "100-1000"
  elif events >= 1000 and events < 3000:
    return "1000-3000"
  elif events >= 3000 and events < 5000:
    return "3000-5000"
  elif events >= 5000 and events < 10000:
    return "5000-10000"
  else:
    return "10000<="

harv_ph_catg["nevents"] = harv_ph_catg["nevents"].apply(lambda events: map_events(events))
harv_ph_catg.head()
```

```python
def map_videos(videop):
  if videop >= 0 and videop < 100:
    return "0-100"
  elif videop >= 100 and videop < 1000:
    return "100-1000"
  elif videop >= 1000 and videop < 3000:
    return "1000-3000"
  elif videop >= 3000 and videop < 5000:
    return "3000-5000"
  elif videop >= 5000 and videop < 10000:
    return "5000-10000"
  else:
    return "10000<="

harv_ph_catg["nplay_video"] = harv_ph_catg["nplay_video"].apply(lambda videop: map_videos(videop))
harv_ph_catg.head()
```

```python
def map_chapters(chapters):
  if chapters >= 0 and chapters < 5:
    return "0-5"
  elif chapters >= 5 and chapters < 10:
    return "5-10"
  else:
    return "10<="

harv_ph_catg["nchapters"] = harv_ph_catg["nchapters"].apply(lambda chapters: map_chapters(chapters))
harv_ph_catg.head()
```

```python
def map_days(days):
  if days >= 0 and days < 10:
    return "0-10"
  elif days >= 10 and days < 30:
    return "10-30"
  elif days >= 30 and days < 50:
    return "30-50"
  else:
    return "50<="

harv_ph_catg["ndays_act"] = harv_ph_catg["ndays_act"].apply(lambda days: map_days(days))
harv_ph_catg.head()
```

#### tiek veidota balansēta datu kopa, izmantojot undersampling tehniku

```python
harv_ph_catg_passes = harv_ph_catg.loc[harv_ph_catg['final_mark'] == 'Pass']
harv_ph_catg_fails = harv_ph_catg.loc[harv_ph_catg['final_mark'] == 'Fail']
harv_ph_catg_fails_sample = harv_ph_catg_fails.sample(700)
harv_ph_catg_subset = pd.concat([harv_ph_catg_passes, harv_ph_catg_fails_sample])
print(harv_ph_catg_subset.shape)
harv_ph_catg_subset.head()
```
